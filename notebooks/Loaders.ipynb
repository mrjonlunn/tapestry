```python
# Load csv file types
def loadCSV(filePath, header, delimiter, multiLine, customSchema, escape):

    try:
        df = spark.read.format('csv') \
            .option("header", header) \
            .option("delimiter", delimiter) \
            .option("multiline", multiLine) \
            .option("escape", escape) \
            .schema(customSchema) \
            .load(filePath) \
            .selectExpr("*", "current_timestamp() AS RowCreatedDate", "current_timestamp() AS RowUpdatedDate", "_metadata.file_name AS SourceFilename")            
        return(df)
    except: 
        moveLandingFile(filePath, 'Failure')
```
