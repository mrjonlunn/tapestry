### Loaders
You can add file loaders here. Using the if statement in the main Raw_Processing notebook, direct them to the write loader using the file extension

```python
# Load csv file types
def loadCSV(file_path, config_fileOptions_header, config_fileOptions_delimiter, config_fileOptions_multiLine, config_fileOptions_escape):

    try:
        df = spark.read.format('csv') \
            .option("header", config_fileOptions_header) \
            .option("delimiter", config_fileOptions_delimiter) \
            .option("multiline", config_fileOptions_multiLine) \
            .option("escape", config_fileOptions_escape) \
            .load(file_path) \
            .selectExpr("*", "current_timestamp() AS MetaCreatedDate", "current_timestamp() AS MetaUpdatedDate", "_metadata.file_name AS MetaSourceFilename")            
        return(df)
    except: 
        moveLandingFile(file_path, 'Failure')
```
